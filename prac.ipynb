```python
from nltk.corpus import stopwords;

sarc_tweets['type'] = 'sarcastic';
gen_tweets['type'] = 'genuine';

import re
def review_to_words(raw_tweets):
    # Remove non-letters
    letters_only = re.sub("[^a-zA-Z@#]", " ", raw_tweets)
    letters_only = letters_only.replace('sarcasm','')
    
    #split individual words by converting in lowercase
    words = letters_only.split()
    
    # converting stop words to set for fast searching
    stops = set(stopwords.words("english"))

    
    # removing stop words
    meaningful_words = [w for w in words if not w in stops]
    
    # joining the words back into one string separated by space and return
    return(" ".join(meaningful_words))
```

```python
#cleaning sarc_tweets and gen_tweets;
num_of_tweetss = sarc_tweets[0].size
# list to hold clean tweets
clean_tweets_sarcastic = []

for i in range(0, num_of_tweetss):
    #clean tweets
    clean_tweets_sarcastic.append(review_to_words(sarc_tweets[0][i]))

num_of_tweetsg = gen_tweets[0].size
clean_tweets_genuine = [];

for i in range(0, num_of_tweetsg):
    clean_tweets_genuine.append(review_to_words(gen_tweets[0][i]))
    
sarcastic_tweets_parsed = [n.replace('sarcasm','') for n in clean_tweets_sarcastic]    
print(clean_tweets_genuine)
```

```python
from nltk.stem.porter import *
ps = PorterStemmer()
sarcastic_tweets_stemmed = [];

for tweet in range(0,len(sarcastic_tweets_parsed)):
    newTweet = []
    for word in sarcastic_tweets_parsed[tweet].split(" "):
        newTweet.append(ps.stem(word));
    sarcastic_tweets_stemmed.append(" ".join(newTweet));
```

```python
from nltk.stem.porter import *
ps = PorterStemmer()
genuine_tweets_stemmed = [];

for tweet in range(0,len(clean_tweets_genuine)):
    newTweet = []
    for word in clean_tweets_genuine[tweet].split(" "):
        newTweet.append(ps.stem(word));
    genuine_tweets_stemmed.append(" ".join(newTweet));
```
